```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('athlete_events.csv')
df_region = pd.read_csv('noc_regions.csv')
df.head()
```

```python
df_region
```

```python
df.shape
```

```python
df = df.merge(df_region, on = 'NOC', how= 'left')
```

```python
df
```

```python
df['Medal'].value_counts()
```

```python
medals = pd.get_dummies(df['Medal'])
df = pd.concat([df,medals], axis =1)
df
```

```python
df = df[df['Season']=='Summer']
```

```python
df
```

```python
df['region'].unique().shape
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.shape
```

```python
df.groupby('NOC').sum()[['Gold', 'Silver', 'Bronze']].sort_values('Gold', ascending = False).reset_index()
```

```python
medal_tally = df.drop_duplicates(subset = ['Team','NOC', 'Games','Year','City','Sport','Event','Medal'])
```

```python
medal_tally.groupby('NOC').sum()[['Gold', 'Silver', 'Bronze']].sort_values('Gold', ascending = False).reset_index()
```
